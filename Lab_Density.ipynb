Метод Парсена-Разенблата для нормального распрдеелния и гаусовского ядра 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.stats import gaussian_kde
```

```python
data = np.random.normal(size=1000)
```

LOO - метод скользящего контроля для нахождения оптимального размера окна

```python
kde = gaussian_kde(data)
bandwidth = kde.scotts_factor() * np.std(data)

# Generate the x-range and calculate the density estimate
x_range = np.linspace(data.min() - 3*bandwidth, data.max() + 3*bandwidth, 100)
```

```python
density = kde(x_range)
```

```python
for x_i in data:
    kernel_values = norm.pdf(x_range, x_i, bandwidth)
    density += kernel_values
density /= data.size * bandwidth
```

```python
plt.plot(x_range, density)
plt.xlabel('x')
plt.ylabel('Density')
plt.show()
```
